```python
#importing libraries
import csv
import pandas as pd
```

```python
#reading the previously saved Toronto2 to a Pandas dataframe and reseting the index
a = pd.read_csv("Toronto2.csv", index_col=0)
a.reset_index(drop=True)
```

```python
#reading the previously downloaded Geospatial_Coordinates file to a Pandas dataframe 
b = pd.read_csv("Geospatial_Coordinates.csv")
b
```

```python
#renaming the coloumn to be consistent with both csv files
b.rename(columns={'Postal Code': 'Postcode'}, inplace=True)

#merging the dataframes on the Postcode column and write to Toronto3.csv file
merged = a.merge(b, on='Postcode')
merged.to_csv("Toronto3.csv")
merged
```
